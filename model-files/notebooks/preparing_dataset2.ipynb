```python
import pandas as pd
import numpy as np
```

```python
import patoolib

patoolib.extract_archive('df1.rar', outdir='.')
patoolib.extract_archive('df2.rar', outdir='.')
```

```python
import pickle

with open('df1.pkl', 'rb') as f:
 df1 = pickle.load(f)

with open('df2.pkl', 'rb') as f:
 df2 = pickle.load(f)
```

```python
df1
```

```python
df2
```

```python
# Replace '\N' with np.nan
df1['primaryProfession'] = df1['primaryProfession'].replace(r'\N', np.nan)
df1['knownForTitles'] = df1['knownForTitles'].replace(r'\N', np.nan)

# Remove all rows where 'primaryProfession' or 'knownForTitles' is NaN
df1 = df1.dropna(subset=['primaryProfession', 'knownForTitles'])
```

```python
# removing all the profession entries that dont inculde either actor or director
def check_profession(row):
  professions = row['primaryProfession'].split(',')
  return 'actor' in professions or 'director' in professions

mask = df1.apply(check_profession, axis=1)

df1 = df1[mask]
```

```python
df1
```

```python
# Delete rows where 'combined' contains '(tvEpisode)', '(tvSeries)', or '(short)'
df2 = df2[~df2['combined'].str.contains('(\(tvEpisode\)|\(tvSeries\)|\(short\))')]
```

```python
df2
```

```python
# Create a mapping dictionary from df2
mapping_dict = pd.Series(df2.combined.values, index=df2.tconst).to_dict()
```

```python
len(mapping_dict)
```

```python
mapping_dict.get('tt0050976')
```

```python
# Replace values in df1
df1['knownForTitles'] = df1['knownForTitles'].apply(lambda x: [mapping_dict.get(str(i)) for i in x.split(',')])
```

```python
df1
```

```python
# Remove rows where 'knownForTitles' contains a None value
df1 = df1[~df1['knownForTitles'].apply(lambda x: any(i is None for i in x))]
```
